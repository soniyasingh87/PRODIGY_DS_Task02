```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/gender_submission.csv")
df1=pd.read_csv("/content/test.csv")
df2=pd.read_csv("/content/train.csv")
df
df1
df2
```

```python
data=[df,df1,df2]
```

```python
data=pd.concat([df,df1,df2])
```

```python
data.info()
```

```python
data.shape
```

```python
data.size
```

```python
data.describe
```

```python
data.hist()
```

```python
data.dropna(subset=["Embarked"], inplace=True)
data["Cabin"].fillna("Unknown", inplace=True)
data["Age"].fillna(data["Age"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
df.duplicated().sum()
df1.duplicated().sum()
df2.duplicated().sum()
```

```python
plt.figure(figsize=(4,2))
sns.histplot(data["Age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Survived")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="Age",y="Pclass",marker="+",palette="inferno",hue="Survived",data=data)
plt.title("Scatter plot of Pclass vs Age")
plt.xlabel("Age")
plt.ylabel("Pclass")
plt.legend(title="Survived")
plt.show()
```

```python
sns.heatmap(data=data.corr(),cmap="RdBu_r",annot=True)
```
